```python
covid_df.drop(["Sno","Time","ConfirmedIndianNational","ConfirmedForeignNational",],inplace=True,axis=1)
```

```python
covid_df.head()
```

```python
covid_df["Date"]=pd.to_datetime(covid_df["Date"],format='%Y-%m-%d')
```

```python
covid_df.head()
```

```python
# for active cases

covid_df['Active_cases']=covid_df['Confirmed']-(covid_df['Cured']+covid_df['Deaths'])
covid_df.tail()
```

```python
covid_df.tail(10)
```

```python
statewise=pd.pivot_table(covid_df,values=['Confirmed','Deaths','Cured'],index="State/UnionTerritory",aggfunc=max)
```

```python
statewise["Recovered rate"]=statewise["Cured"]*100/statewise["Confirmed"]
```

```python
statewise["Mortality rate"] = statewise["Deaths"] * 100 / statewise["Confirmed"]
```

```python
statewise=statewise.sort_values(by="Confirmed",ascending=False)
```

```python
statewise.style.background_gradient(cmap="cubehelix")
```

```python
#top 10 active cases states
top_10_active_cases_states=covid_df.groupby(by='State/UnionTerritory').max()[['Active_cases','Date']].sort_values(by=['Active_cases'],ascending=False).reset_index()
```

```python
fig=plt.figure(figsize=(20,8))
```

```python
plt.title("top_10_active_cases_states",size=35)
```

```python
ax=sns.barplot(data=top_10_active_cases_states.iloc[:15],y="Active_cases",x="State/UnionTerritory",linewidth=4,edgecolor='blue')
```

```python
top_10_active_cases_states = covid_df.groupby(by='State/UnionTerritory').max()[['Active_cases', 'Date']].sort_values(
    by=['Active_cases'], ascending=False).reset_index()
fig=plt.figure(figsize=(20,10))
plt.title("top_10_active_cases_states",size=35)
ax=sns.barplot(data=top_10_active_cases_states.iloc[:15],y="Active_cases",x="State/UnionTerritory",linewidth=4,edgecolor='blue')
plt.xlabel("States")
plt.ylabel("total active cases")
plt.show()
```

```python
#top states with deaths
top_10_death_states=covid_df.groupby(by='State/UnionTerritory').max()[['Deaths','Date']].sort_values(by=['Deaths'],ascending=False).reset_index()
fig=plt.figure(figsize=(15,8))
plt.title("top_10_death_states",size=25)
ax=sns.barplot(data=top_10_death_states.iloc[:10],y="Deaths",x="State/UnionTerritory",linewidth=3,edgecolor='green')
plt.xlabel("States")
plt.ylabel("total death cases")
plt.show()
```
